```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv("bench.csv")
data
```

```python
data.shape
```

```python
data_copy = data.copy(deep=True) # Make a deep copy, including a copy of the data and the indices. 
```

```python
# E: how many years of education
# M: How many hours they work per week
# O: Do they earn more or less than $ 50000 a year? (I'll try to guess that column) # classification problem(0/1)

data = data.drop("A", axis=1) # axis: Whether to drop labels from the index (0 or ‘index’) or columns (1 or ‘columns’).
data = data.drop("B", axis=1)
data = data.drop("C", axis=1)
data = data.drop("D", axis=1)
data = data.drop("F", axis=1)
data = data.drop("G", axis=1)
data = data.drop("H", axis=1)
data = data.drop("I", axis=1)
data = data.drop("J", axis=1)
data = data.drop("K", axis=1)
data = data.drop("L", axis=1)
data = data.drop("N", axis=1)
data
```

```python
data = pd.get_dummies(data)
labels = np.array(data["O"])  # O: target
data = data.drop("O", axis=1)
```

```python
train_x, test_x, train_y, test_y = train_test_split(data, labels, test_size=0.2, random_state=42)
print("Train_x shape:", train_x.shape)
print("Test_x shape:", test_x.shape)
print("Train_y shape:", train_y.shape)
print("Test_y shape:", test_y.shape)
```

```python
data_copy = pd.get_dummies(data_copy)
labels_full = np.array(data_copy["O"])
data_copy = data_copy.drop("O", axis=1)
```

```python
ftrain_x, ftest_x, ftrain_y, ftest_y = train_test_split(data_copy, labels_full, test_size=0.2, random_state=42)
print("ftrain_x shape:", ftrain_x.shape)
print("ftest_x shape:", ftest_x.shape)
print("ftrain_y shape:", ftrain_y.shape)
print("ftest_y shape:", ftest_y.shape)
```

```python
data.head()
```

```python
data_copy.head()
```

```python
knn = KNeighborsClassifier(n_neighbors=5, metric="euclidean")
knn.fit(train_x, train_y)
y_predict = knn.predict(test_x)

test_x["target"] = test_y
test_x["pred"]   = y_predict

print(confusion_matrix(test_y, y_predict))
print(roc_auc_score(test_y, y_predict))
```

```python
sns.scatterplot(x="E", y="M", hue="target", data=test_x)
# real data
```

```python
sns.scatterplot(x="E", y="M", hue="pred", data=test_x)
# predictions
```

```python
test_x = test_x.drop("pred", axis=1)
test_x = test_x.drop("target", axis=1)

knn1 = KNeighborsClassifier(n_neighbors=10, metric="euclidean")
knn1.fit(train_x, train_y)

y_predict2 = knn1.predict(test_x)

test_x["target"] = test_y
test_x["pred"] = y_predict2

print(confusion_matrix(test_y, y_predict2))
print(roc_auc_score(test_y, y_predict2))
```

```python
sns.scatterplot(x="E", y="M", hue="pred", data=test_x)
```

```python
knn_full = KNeighborsClassifier(n_neighbors=5, metric="euclidean")
knn_full.fit(ftrain_x, ftrain_y)

y_predict_full = knn_full.predict(ftest_x)

ftest_x["target"] = ftest_y
ftest_x["pred"] = y_predict_full

print(confusion_matrix(ftest_y, y_predict_full))
print(roc_auc_score(ftest_y, y_predict_full))
```

```python
ftest_x = ftest_x.drop("target", axis=1)
ftest_x = ftest_x.drop("pred", axis=1)

knn_full2 = KNeighborsClassifier(n_neighbors=10, metric="euclidean")
knn_full2.fit(ftrain_x, ftrain_y)

y_predict_full2 = knn_full2.predict(ftest_x)

ftest_x["target"] = ftest_y
ftest_x["pred"] = y_predict_full2

print(confusion_matrix(ftest_y, y_predict_full2))
print(roc_auc_score(ftest_y, y_predict_full2))
```

```python
ftest_x = ftest_x.drop("target", axis=1)
ftest_x = ftest_x.drop("pred", axis=1)

knn_full3 = KNeighborsClassifier(n_neighbors=40, metric="euclidean")
knn_full3.fit(ftrain_x, ftrain_y)

y_predict_full3 = knn_full3.predict(ftest_x)

ftest_x["target"] = ftest_y
ftest_x["pred"] = y_predict_full3

print(confusion_matrix(ftest_y, y_predict_full3))
print(roc_auc_score(ftest_y, y_predict_full3))

# Curse of Dimensionality
```
