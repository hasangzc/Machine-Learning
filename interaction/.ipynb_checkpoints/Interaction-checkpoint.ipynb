```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols
from scipy import stats
```

```python
data = pd.read_csv("Credit.csv")
data.head()
```

```python
data.shape
```

```python
data = pd.get_dummies(data)
```

```python
labels = np.array(data["Balance"])
data = data.drop("Balance", axis=1)
data.head()
```

```python
data = sm.add_constant(data)
reg = sm.OLS(labels, data)
results  = reg.fit()
print(results.summary())
```

```python
data["Interaction"] = data["Education"] * data["Gender_Female"]
reg = sm.OLS(labels, data)
results = reg.fit()
print(results.summary())
```

```python
data["Interaction2"] = data["Income"] * data["Student_Yes"]
reg = sm.OLS(labels, data)
results = reg.fit()
print(results.summary())
```

```python
data["Interaction3"] = data["Income"] * data["Ethnicity_African American"]
reg = sm.OLS(labels, data)
results = reg.fit()
print(results.summary())
```

```python
data["IncomeSq"] = data["Income"] * data["Income"] 
reg = sm.OLS(labels, data)
results = reg.fit()
print(results.summary())
```

```python
data["Incomecube"] = data["Income"] * data["Income"] * data["Income"] 
reg = sm.OLS(labels, data)
results = reg.fit()
print(results.summary())
```

```python
data = data.drop("Interaction", axis=1)
data = data.drop("Interaction2", axis=1)
data = data.drop("Interaction3", axis=1)
data = data.drop("IncomeSq", axis=1)
data = data.drop("Incomecube", axis=1)
```

```python
reg = sm.OLS(labels, data)
results = reg.fit()
print(results.summary())
```
